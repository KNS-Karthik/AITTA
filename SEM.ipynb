#Regularized Versions

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_deriv(x):
    return x * (1 - x)
np.random.seed(0)
X = np.random.rand(100, 2)
y = ((X.sum(axis=1)) > 1).astype(int).reshape(-1, 1)
W1, b1 = np.random.rand(2, 4), np.zeros((1, 4))
W2, b2 = np.random.rand(4, 1), np.zeros((1, 1))
for epoch in range(100):
    a1 = sigmoid(X @ W1 + b1)
    a2 = sigmoid(a1 @ W2 + b2)
    d2 = (a2 - y) * sigmoid_deriv(a2)
    d1 = d2 @ W2.T * sigmoid_deriv(a1)
    W2 -= a1.T @ d2 * 0.1
    b2 -= d2.sum(axis=0, keepdims=True) * 0.1
    W1 -= X.T @ d1 * 0.1
    b1 -= d1.sum(axis=0, keepdims=True) * 0.1
    if epoch % 10 == 0:
        loss = np.mean(-y * np.log(a2 + 1e-8) - (1 - y) * np.log(1 - a2 + 1e-8))
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
print("Accuracy:", np.mean((a2 > 0.5) == y))
```

#Unregularized

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    return x * (1 - x)

np.random.seed(0)
X = np.random.rand(100, 2)
y = ((X.sum(axis=1)) > 1).astype(int).reshape(-1, 1)

W1 = np.random.rand(2, 4)
b1 = np.zeros((1, 4))
W2 = np.random.rand(4, 1)
b2 = np.zeros((1, 1))

learning_rate = 0.1

for epoch in range(100):
    a1 = sigmoid(X @ W1 + b1)
    a2 = sigmoid(a1 @ W2 + b2)

    d2 = (a2 - y) * sigmoid_deriv(a2)
    d1 = d2 @ W2.T * sigmoid_deriv(a1)

    W2 -= learning_rate * (a1.T @ d2)
    b2 -= learning_rate * d2.sum(axis=0, keepdims=True)
    W1 -= learning_rate * (X.T @ d1)
    b1 -= learning_rate * d1.sum(axis=0, keepdims=True)

    if epoch % 10 == 0:
        loss = np.mean(-y * np.log(a2 + 1e-8) - (1 - y) * np.log(1 - a2 + 1e-8))
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

print("Accuracy:", np.mean((a2 > 0.5) == y))
```

#cats and dogs

```python
import os
import zipfile
import requests
import shutil
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models

url = "https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip"
filename = "cats_and_dogs_filtered.zip"

if not os.path.exists(filename):
    r = requests.get(url, stream=True)
    with open(filename, "wb") as f:
        shutil.copyfileobj(r.raw, f)

with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall()

base_dir = "cats_and_dogs_filtered"
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')

train_gen = ImageDataGenerator(rescale=1.0/255)
val_gen = ImageDataGenerator(rescale=1.0/255)

train_data = train_gen.flow_from_directory(
    train_dir, target_size=(150, 150), batch_size=20, class_mode='binary'
)
val_data = val_gen.flow_from_directory(
    val_dir, target_size=(150, 150), batch_size=20, class_mode='binary'
)

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(train_data, epochs=5, validation_data=val_data)

plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.title("Accuracy")
plt.legend()
plt.show()
```
